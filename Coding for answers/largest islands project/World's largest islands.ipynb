This dataset `largest-islands.csv` contains information about the 100 largest islands in the world. The unit for the `area` column is km2.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('largest-islands.csv')
df.head()
```

### Project Ideas

- What are the 10 largest islands in the tropics?

- What are the largest islands in each `region`?

- Create a line graph with `area` on the y-axis and `rank` on the x-axis. The data should be ordered by `rank`, from largest to smallest.

- What islands are composed of multiple countries?
	- Hint: try `countries.str.contains(",")`

```python
# Solution 1
# The 10 largest island.
largest_islands = df.sort_values('area', ascending=False).head(10)
largest_islands
```

```python
# Solution 2
# largest island by region.
by_region = df.sort_values('area', ascending=False).groupby('region')
by_region.first()
```

```python
# Solution 3
# Ploting line graph ordered by rank with rank on x-axis and area on y-axis.
for_graph = df.sort_values('rank', ascending=False)
plt.plot(for_graph['rank'], for_graph['area'])
plt.show()
```

```python
# Solution 4
# Islands composed of multiple countries.
multi_countries = df[df['countries'].str.contains(',')]
multi_countries
```
