# Jean pockets

Are women jean pockets ridiculously small? Let's look at the data!

The file `jean-pocket-measurements.csv` shows pocket measurements for 20 popular brands.
Four pairs of jeans from each brand were measured: men's and women's skinny and straight styles. All jeans were designated a 32-inch waistband.

The following measurements are included:
- `height_front`: maximum front pocket height (cm)
- `height_back`: maximum back pocket height (cm)
- `width_front`: maximum front pocket width (cm)
- `width_back`: maximum back pocket width (cm)

```python
import pandas as pd
df = pd.read_csv('jean-pocket-measurements.csv')
df.head()
```

### Project Ideas

- What is the average difference in pocket `height_front` between women's and men's jeans?

- Is there a significant difference in pocket `height_front` between skinny and straight styles within the same gender?

- How do back pocket sizes compare between women's and men's jeans?

- Measure your phone's height. What percentage of women's and men's jeans can comfortably fit your phone in the pockets?



```python
# Solution 1
phw = df.query('gender == "women"')['height_front'].mean()
phm = df.query('gender == "men"')['height_front'].mean()
print(f'The difference b/w pocket height_front of men and women is: {round(phfm-phfw,2)} cm')
```

```python
# Difference in pocket height_front between skinny and straight styles within (Women)
skw = df.query('gender == "women" and style == "skinny"')['height_front'].mean()
stw = df.query('gender == "women" and style == "straight"')['height_front'].mean()

print(f'The difference b/w pocket height_front of straight women and skinny women is: {round(stw-skw,3)} cm')

# difference in pocket height_front between skinny and straight styles within (Men)
skm = df.query('gender == "men" and style == "skinny"')['height_front'].mean()
stm = df.query('gender == "men" and style == "straight"')['height_front'].mean()

print(f'The difference b/w pocket height_front of skinny men and straight men is: {round(skm-stm,3)} cm')
```

```python
# solution 3
# Women pocket size
bphw = df.query('gender == "women"')['height_back'].mean()
bpww = df.query('gender == "women"')['width_back'].mean()
bpocketsize_women = (bphw*bpww)
print(f'The pocket size of women is: {round(bpocketsize_women,2)} sqcm')

# Men pocket size
bphm = df.query('gender == "men"')['height_back'].mean()
bpwm = df.query('gender == "men"')['width_back'].mean()
bpocketsize_men = (bphm*bpwm)
print(bpocketsize_men)
print(f'The pocket size of men is: {round(bpocketsize_men,2)} sqcm')

# Difference between pocket size of Men and Women.
print(f'The difference b/w pocket size of men and women is: {round(bpocketsize_men-bpocketsize_women,2)} sqcm')
print(f"The Pocket size of men jeans is greater then women by: '{round(bpocketsize_men-bpocketsize_women)} sqcm'")
```
