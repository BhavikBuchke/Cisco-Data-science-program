# People on Banknotes

Whose faces appear on banknotes?

The file `people-on-banknotes.csv` contains data about individuals featured on banknotes from 38 countries. This dataset spans all 22 subregions and sub-subregions of the world, as defined by the United Nations Statistics Division's geoscheme.

It profiles 241 people, detailing their occupations and the year they first appeared on a banknote. Additionally, it includes their year of death — or `NaN` if they were still alive when the dataset was compiled.

Most banknotes were issued after the featured individual’s death. The column `first_death_diff` calculates the difference between the year of their first appearance on a banknote and their year of death (or remains `NaN` if the person was still living at the time of curation).

```python
import pandas as pd

df = pd.read_csv('people-on-banknotes.csv')
df
```

### Quick cleaning

The same person can appear on multiple banknotes. Below we drop the `value` column and remove duplicate people.

```python
df = df.drop(columns=['value'])
df = df.drop_duplicates(subset="name")
df
```

### Project Ideas

```python
#Solution 1
# Proportion of individuals featured are male versus female.
propotion = df['gender'].value_counts(normalize=True)
propotion.round(2)
```

```python
print(df['occupation'].unique())
```

```python
# Solution 2
# More commonly depicted Writer or Politician
writer = df.query('occupation == "Writer"').value_counts().sum()
politicians = df.query('occupation == "Politician"').value_counts().sum()

print(f"Number of Writers: {writer}")
print(f"Number of Politicians: {politicians}")

if writer > politicians:
    print("Writers are more commonly depicted.")
elif politicians > writer:
    print("Politicians are more commonly depicted.")
else:
    print("Writers and Politicians are depicted equally.")
```

```python
# Solution 3
# Featured individuals are musicians
musician = df.query('occupation == "Musician"').value_counts().sum()
print(f'Number of Mucisians: {musician}')

# Percentage of featured individuals are musicians
percent_musician = (musician/len(df))*100
print(f'Percentage of mucisian: {round(percent_musician,2)}')
```

```python
# Solution 4
# percentage of banknotes were issued before the person’s death.
issued_before_death = df.query('first_death_diff < 0' or 'first_death_diff == NaN').value_counts().sum()
print(f'The number of banknotes issued before the person’s death: {issued_before_death}')
percent_before_death = (issued_before_death/len(df))*100
print(f'Percentage of banknotes issued before the person’s death: {round(percent_before_death,2)}')
```

```python
# Solution 5
# The oldest historical figure in the dataset.
oldest_figure = df.query('first_appearance == first_appearance.max()')
oldest_figure
```

```python
# Solution 6
# Countries that feature the oldest historical figures on their banknotes
df['death'] = pd.to_numeric(df['death'])
oldest_figure_by_country = df.groupby('country')['death'].median().sort_values()
print(f' Top 10 countries that feature the oldest historical figures on their banknotes:\n{oldest_figure_by_country.head(10)}')
```

```python
# Solution 7
# Percentage of individuals died at least 100 years before appearing on a banknote.
died_at_least_100 = df.query('first_death_diff >= 100').value_counts().sum()
print(f'The number of individuals died at least 100 years before appearing on a banknote: {died_at_least_100}')
percent_least_100 = (died_at_least_100/len(df))*100
print(f'Percentage of individuals died at least 100 years before appearing on a banknote: {round(percent_least_100,2)}')
```

```python
# Solution 8
#  Individuals appeared on a banknote just one year after their death.
after_one_year = df.query('first_death_diff == 1.0').value_counts().sum()
print(f'The number of individuals appeared on a banknote just one year after their death: {after_one_year}')
```
