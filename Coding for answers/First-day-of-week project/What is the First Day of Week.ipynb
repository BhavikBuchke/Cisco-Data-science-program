# What is the First Day of the Week?

Whether the Gregorian calendar shows Sunday or Monday as the first day of the week depends on where you live.

Do more countries start the week on Sunday or Monday? What about people? What about by continent?

The file `first-day-of-week.csv` shows the first day of the week for each territory. The file `population.csv` shows the population in the year 2020 for each territory in millions, and the file `four-regions.csv` specifies whether each territory is in asia, europe, africa, or the americas.

```python
import pandas as pd
df = pd.read_csv('first-day-of-week.csv')
df.head()
```

```python
pop = pd.read_csv('population.csv')
pop.head()
```

```python
regions = pd.read_csv('four-regions.csv')
regions.head()
```

### Project Ideas

- How many territories show Friday, Saturday, Sunday, and Monday as the `first_day` of the week?

- How many people start the week on Friday, Saturday, Sunday, and Monday?
	- Hint: This will involve a `merge`.

- Which of the `four_regions` predominantly start the week on Sunday? On Monday? Are there any regions that are more divided between Sunday and Monday?
	- Hint: This will also involve a `merge`.

```python
# Solution 1
# Territories show Friday as first day of week.
fri_first_day_count = df.query('first_day == "fri"').shape[0]
sat_first_day_count = df.query('first_day == "sat"').shape[0]
sun_first_day_count = df.query('first_day == "sun"').shape[0]
mon_first_day_count = df.query('first_day == "mon"').shape[0]

print(f"Number of territories where first day is->\nFriday: {fri_first_day_count}\nSaturday:{sat_first_day_count}\nSunday:{sun_first_day_count}\nMonday:{mon_first_day_count}")
```

```python
# Solution 2
# Creating a new dataframe by merging both df and pop.
merged_df = pd.merge(df, pop, on='alpha3')
merged_df.head()
```

```python
# People start the week on Friday, Saturday, Sunday, and Monday.
friday_as_firstday = merged_df.query('first_day == "fri"')['population'].sum()
saturday_as_firstday = merged_df.query('first_day == "sat"')['population'].sum()
sunday_as_firstday = merged_df.query('first_day == "sun"')['population'].sum()
monday_as_firstday = merged_df.query('first_day == "mon"')['population'].sum()

print(f"Number of people who start the week on Friday: {friday_as_firstday} million")
print(f"Number of people who start the week on Saturday: {round(saturday_as_firstday,2)} million")
print(f"Number of people who start the week on Sunday: {round(sunday_as_firstday,2)} million")
print(f"Number of people who start the week on Monday: {round(monday_as_firstday,2)} million")
```

```python
# Solution 3
# Regions predominantly start the week on Sunday or On Monday.
all_merged = pd.merge(merged_df, regions, on='alpha3')
all_merged.head()
```

```python
# Calculate total territories for each region.
All_regions = all_merged['four_regions'].value_counts()

sunday_regions = all_merged.query('first_day == "sun"')['four_regions'].value_counts()
monday_regions = all_merged.query('first_day == "mon"')['four_regions'].value_counts()

# Combine Sunday and Monday and All regions in one dataframe.
combined = pd.DataFrame({'Sunday_Start': sunday_regions,'Monday_Start': monday_regions,'Total_Territories': All_regions}).fillna(0)

# Calculate percentages
combined['Sunday_Percent'] = round((combined['Sunday_Start'] / combined['Total_Territories']) * 100)
combined['Monday_Percent'] = round((combined['Monday_Start'] / combined['Total_Territories']) * 100)

print(f"Distribution of First Day of the Week by Region:\n{combined}")
```

```python
# For regions that are more divided into Sunday and Monday.
print("Regions that are divided between Sunday and Monday:")

for region, row in combined.iterrows():
    if (row['Sunday_Percent'] > 30 and row['Monday_Percent'] > 30):
        print(f"\n{region}: Sunday {round(row['Sunday_Percent'])}% ({int(row['Sunday_Start'])} territories), Monday {round(row['Monday_Percent'])}% ({int(row['Monday_Start'])} territories)")
    elif abs(row['Sunday_Percent'] - row['Monday_Percent']) < 25 and (row['Sunday_Percent'] > 10 and row['Monday_Percent'] > 10):
        print(f"{region}: Sunday {round(row['Sunday_Percent'])}% ({int(row['Sunday_Start'])} territories), Monday {round(row['Monday_Percent'])}% ({int(row['Monday_Start'])} territories)")

print("\nAsia and Americas show a more balanced distribution between Sunday and Monday as the first day of the week as compare to Europe and Africa.")
```
